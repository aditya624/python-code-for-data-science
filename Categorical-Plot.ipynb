```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(data = tips, x = 'sex', y = 'total_bill')
```

```python
sns.barplot(data = tips, x = 'time', y = 'total_bill')
```

```python
sns.countplot(x = 'sex', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow')
```

```python
sns.boxplot(data = tips, palette = 'rainbow', orient = 'h')
```

```python
sns.boxplot(data = tips, palette = 'coolwarm', x = 'day', y = 'total_bill', hue = 'smoker')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow')
```

```python
sns.violinplot(data = tips, x = 'day', y = 'total_bill', palette = 'Set1', hue = 'sex')
```

```python
sns.violinplot(data = tips, x = 'day', y = 'total_bill', palette = 'Set1', hue = 'sex', split = True)
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill')
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill', jitter = False)
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', palette = 'Set1')
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', palette = 'Set1', dodge = True)
```

```python
sns.swarmplot(data = tips, x = 'day', y = 'total_bill')
```

```python
sns.swarmplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex')
```

```python
sns.swarmplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', dodge = True)
```

```python
sns.violinplot(data = tips, x = 'tip', y = 'day', palette = 'rainbow')
sns.swarmplot(data = tips, x = 'tip', y = 'day', color = 'black', size = 3)
```

```python
sns.catplot(x = 'sex', y = 'total_bill', data = tips, kind = 'bar')
```
