```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', palette = 'coolwarm')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', palette = 'coolwarm', markers = ['o', 'v'], scatter_kws = {'s':100})
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, row = 'time', col = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', palette = 'coolwarm', hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', palette = 'coolwarm', hue = 'sex',  aspect = 0.6, height = 10)
```
